```python
import matplotlib.pyplot as plt
x=range(32)
y1=cars["drat"]
plt.plot(x,y1,color='red',linewidth=.2)  #this is area plot normal boundary
plt.stackplot(x,y1,color='purple',alpha=.3)#this fills the background of upper graph
y2=cars["wt"]
plt.plot(x,y2,color="yellow",alpha=.6)  #similarly for graph 2
plt.stackplot(x,y2,color="green",alpha=.45)
plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt
import numpy as np
label=['PYTHON','C','C#','FLUTTER']
sizes=[80,50,30,40]
plt.pie(sizes,labels=label,autopct='%.1f',shadow=True,explode=(0,1,.1,0),startangle=90) #observe the format it will work as it is
plt.title('programmming languages') #labels is used to assign label mentioned..labels is keyword,autopct is used for shoing mean percentahe
plt.show() #explode will explode out the wedge with '1'assign.start angle will rotatte the pie chart by 90 degree (see from starting ie python)
```

```python
#DOUGHNUT CHART --ITS BASICALLY THE COMBINATION OF TWO PIE CHARTS..NOTE THE RADIUS MENTIONED IN TWO PIE CHARTS..NOTE SIZE CENTRE
import matplotlib.pyplot as plt
%matplotlib inline
label=['python','flutter','c','java']
sizes=[80,50,30,40]
size_centre=[3]
pie1=plt.pie(sizes,labels=label,autopct='%.1f',shadow=True,radius=1.5)
pie2=plt.pie(size_centre,radius=1.1,colors='y',shadow=True)
plt.show()

```
